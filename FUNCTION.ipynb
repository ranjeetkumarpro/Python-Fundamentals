```python
def even_odd(num):
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")
```

```python
#write function to swap two number without using third variable
def swap(a,b):
    a,b=b,a
    print(a)
    print(b)
```

```python
swap(3,5)
```

```python
swap(3,7)
```

```python
def swap(a,b):
    temp=a
    a=b
    b=temp
    print(a)
    print(b)
```

```python
swap(3,8)
```
